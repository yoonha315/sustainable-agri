# Urban Heat Island Effect and Green Spaces Data Analysis

```python
import sqlite3
import pandas as pd
```

```python
# Connect to the SQLite database
conn = sqlite3.connect('uhi_green.db')

# Get a cursor object
cursor = conn.cursor()
```

## Create Data Frames

**USA Urban Temperature (Cities)**

```python
cities_of_interest = ['Nashville', 'Cincinnati', 'Minneapolis', 'Louisville',
                     'Charlotte', 'Birmingham', 'Philadelphia', 'Albany', 'Grand Rapids',
                      'Jacksonville', 'New Orleans', 'Austin', 'Seattle']
```

To concentrate our analysis on cities that can significantly influence UHI data, we've chosen cities from across the USA. We've chosen them from the top 4, mid 5, and bottom 4 based on the ranking of their green space compared to population density.

These selections are based on data from [Stacker](https://stacker.com/environment/cities-most-green-space-capita), which ranks cities by their green space per capita.

To streamline our dataset, we compute the average of numerical data for each city, resulting in a single row of data for each city.

```python
cities_str = ', '.join(["'{}'".format(city) for city in cities_of_interest])

query = f"""
    SELECT NAME AS City, 
           AVG(SQKM_FINAL) AS SQKM,
           AVG(URB_D_MEAN) AS Urban_Day, 
           AVG(BUF_D_MEAN) AS Buffer_Day, 
           AVG(D_T_DIFF) AS Difference_Day,
           AVG(URB_N_MEAN) AS Urban_Night, 
           AVG(BUF_N_MEAN) AS Buffer_Night,
           AVG(N_T_DIFF) AS Difference_Night
    FROM Global_UHI_UHI
    WHERE ISO3 = 'USA'
    GROUP BY NAME
    HAVING NAME IN ({cities_str})
"""

urban_heat_df = pd.read_sql_query(query, conn)

urban_heat_df
```

**Explanatory Variables**

```python
states_of_interest = ['Tennessee', 'Ohio', 'Minnesota', 'Kentucky',
                      'North Carolina', 'Alabama', 'Pennsylvania', 'New York', 'Michigan',
                     'Florida', 'Louisiana', 'Texas', 'Washington']
```

Above, we've chosen the cities of interest. Since the Green Space and Population data are identified using the state name rather than the city name, we've selected the states corresponding to the cities we've chosen.

**USA Green Space (States)**

```python
states_str = ', '.join(["'{}'".format(state) for state in states_of_interest])

query = f"""
    SELECT State, 
       AVG(Overall_Rank) AS Overall_Rank,
       AVG(Green_Energy_Rank) AS Green_Energy_Rank,
       AVG(Open_Spaces_Beauty) AS Open_Spaces_Beauty,
       AVG(Waste_Diversion_Recycling) AS Waste_Diversion_Recycling,
       AVG(Racial_Justice_Access_to_Clean_Outdoors) AS Equitable_Outdoor_Access
    FROM USA_park
    WHERE State IN ({states_str})
    GROUP BY State
"""

green_space_data = pd.read_sql_query(query, conn)

green_space_data
```

**USA Population (States)**

```python
states_str = ', '.join(["'{}'".format(state) for state in states_of_interest])

query = f"""
    SELECT [Geographic Area] AS State, Population
    FROM USA_population
    WHERE [Geographic Area] IN ({states_str}) AND Year = 2013
"""

population_data = pd.read_sql_query(query, conn)

population_data
```

**Match States and Cities**

```python
cities_states_data = {
    'City': [
        'Nashville',
        'Cincinnati',
        'Minneapolis',
        'Louisville',
        'Charlotte',
        'Birmingham',
        'Philadelphia',
        'Albany',
        'Grand Rapids',
        'Jacksonville',
        'New Orleans',
        'Toledo',
        'Seattle',
        'Austin'
    ],
    'State': [
        'Tennessee',
        'Ohio',
        'Minnesota',
        'Kentucky',
        'North Carolina',
        'Alabama',
        'Pennsylvania',
        'New York',
        'Michigan',
        'Florida',
        'Louisiana',
        'Ohio',
        'Washington',
        'Texas'
    ]
}
cities_states_df = pd.DataFrame(cities_states_data)
cities_states_df.head()
```

```python
cities_merged_df = pd.merge(cities_states_df, green_space_data, on='State')
cities_merged_df.head()
```

```python
cities_merged_df = pd.merge(cities_states_df, population_data, on='State')
cities_merged_df.head()
```

**USA UHI & Green Space**

```python
uhi_green_merged_df = pd.merge(urban_heat_df, cities_merged_df, on='City')

uhi_green_merged_df
```

**USA UHI & Population**

```python
cities_merged_df = pd.merge(cities_states_df, population_data, on='State')

uhi_pop_merged_df = pd.merge(cities_merged_df, urban_heat_df, on='City')

uhi_pop_merged_df
```

**Final Data Frame**

```python
final_df = pd.merge(uhi_pop_merged_df, green_space_data, on='State')
final_df
```

```python
# Close the connection
conn.close()
```

The Population Density vs. UHI scatter plot demonstrates randomly scattered points, implying that there is no relationship between UHI and population.

### Side by Side Bar Plots: Green Space Comparison

```python
categories = ['Overall_Rank', 'Green_Energy_Rank', 'Open_Spaces_Beauty', 'Waste_Diversion_Recycling', 'Equitable_Outdoor_Access']
colors = ['darkgreen', 'olivedrab', 'forestgreen', 'yellowgreen', 'seagreen']

plt.figure(figsize=(12, 6))

num_states = len(green_space_data)

bar_width = 0.15

index = range(num_states)

for i, category in enumerate(categories):
    plt.bar([x + i * bar_width for x in index], green_space_data[category], bar_width, label=category, color=colors[i])

plt.xlabel('State')
plt.ylabel('Average Rank')
plt.title('Average Ranks by State')
plt.xticks([i + (len(categories) - 1) * bar_width / 2 for i in index], green_space_data['State'])  # Setting x-axis ticks to state names
plt.legend()

plt.tight_layout()
plt.show()
```

I generated a side-by-side bar plot to assess the alignment of other green rankings with the overall ranking. However, upon analysis of this plot, it becomes apparent that there are notable disparities between the rankings in most of states. These differences suggest a weak correlation in the ranking of green spaces.

### Correlation Heat Map

```python
import seaborn as sns
```

```python
reduced_final_df = final_df[['Difference_Day', 'Difference_Night', 'Population', 'Overall_Rank']]

corr = reduced_final_df.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

Now, I have generated a correlation heat map to directly visualize the project question, which aimed to determine if population and green space factors affect the UHI. Prior to plotting this, I narrowed down the variables to 'Difference_Day', 'Difference_Night', 'Population', and 'Overall_Rank' so that we could focus on the variables that might have a greater influence on UHI. From the plot, it is evident that population shows a moderately negative correlation with the temperature difference between urban and buffer areas during the daytime, and a weak negative correlation during nighttime. Regarding green space, both daytime and nighttime temperature differences between urban and buffer areas exhibit weak negative correlations. Despite the weak correlations, we can observe that our assumption holds true: higher rankings in green space and lower population densities correspond to higher temperature differences between urban and buffer area.

### Fit Multiple Linear Regression

```python
import statsmodels.api as sm
```

```python
# Load the data (assuming final_df is already loaded)
# Replace the file path with the actual path to your data

# Define the response variable (y) and explanatory variables (X)
y = final_df["Difference_Day"]
X = final_df[["Population", "Overall_Rank"]]

# Add a constant term to the explanatory variables
X = sm.add_constant(X)

# Fit the multiple linear regression model
model = sm.OLS(y, X).fit()

# Print the summary of the regression results
print(model.summary())
```

Finally, I attempted to fit a least squares model to assess the predictive capability of our data. With an adjusted R squared value of 0.235, it's evident that this model will perform poorly in predicting UHI using the available data. Furthermore, observing that the confidence intervals of the explanatory variables encompass 0 indicates that they have minimal impact on UHI.

## Final Conclusion

In conclusion, our analysis suggests that population density and green space ranking have minimal influence on UHI. This discrepancy may arise from the fact that population and green space data were collected at the state level, whereas UHI data were gathered specifically from urban areas. This difference in data collection likely obscures the true impact of population and green space on UHI, leading to a poorer representation of their influence.
